```python
#importing libraries
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasets/hepatitis/master/data/hepatitis.csv')
#displaying first 5 rows of dataset
df.head(5)
```

```python
#displays rows and columns
df.shape
```

```python
#Select numeric columns
num_col = df.select_dtypes(include=np.number).columns
```

```python
# Display data types of numeric columns
print("\n Data Types of Numeric Columns:")
print(df.dtypes[num_col])
```

```python
# Find the number of missing values in numeric columns before replacement
print("\n Missing Values Before Replacement:")
print(df[num_col].isna().sum())
```

```python
# Replace missing values in numeric columns with column mean
for col in num_col:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
# Check if missing values are removed
print("\n Missing Values After Replacement:")
for col in num_col:
    missing_count = df[col].isna().sum()
    print(f"    {col}: {missing_count} ")
```
